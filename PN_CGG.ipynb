## 5.  Write down a function called "inspect" to inspect array1 and array2 Shape, length, Dimensions, and Datatype using Preetytable. Your pretty table should contain a title and look like similar to the following:


```python
def inspect (array): #array es lo q voy a analizar
    PT=PrettyTable()
    PT.title="Numpy array Inspection"
    PT.field_names=["Shape", "Length", "Dimensions", "DataType"]
    PT.add_row([array.shape, len(array), array.ndim, array.dtype]) #Adding row
    print(PT)
```

```python
def inspect (array, arrayb): #array es lo q voy a analizar
    PT=PrettyTable()
    PT.title="Numpy array Inspection"
    PT.field_names=["Array","Shape", "Length", "Dimensions", "DataType"]
    PT.add_row(["1", array.shape, len(array), array.ndim, array.dtype]) #Adding row
    PT.add_row(["2",arrayb.shape, len(arrayb), arrayb.ndim, arrayb.dtype]) #Adding row
    print(PT)
```

## 8.Create a function called "calculation" to display `sum, min, max, mean, standard deviation, median calculations about a NumPy array using PrettyTable. Test it for array1,2,3


```python
def calculation (array, arraya, arrayb): #array es lo q voy a analizar
    PT=PrettyTable()
    PT.title="Numpy array Calculations"
    PT.field_names=["Sum", "Min", "Max", "Mean", "Std", "Median"]
    PT.add_row([ array.sum(), array.min(), array.max(), array.mean(), np.std(array), np.median(array)]) #Adding row
    PT.add_row([ arraya.sum(), arraya.min(), arraya.max(), arraya.mean(), np.std(arraya), np.median(arraya)]) #Adding row
    PT.add_row([ arrayb.sum(), arrayb.min(), arrayb.max(), arrayb.mean(), np.std(arrayb), np.median(arrayb)]) #Adding row
    print(PT)
```

```python
calculation(array1, array2, array3)
```

## 9.Create a function called "arithmatic" to display `Subtraction (np.subtract(a,b)), Addition,  Multiplication calculations about two NumPy array using PrettyTable. Test it for

```python
def arithmatic (array, arraya): #array es lo q voy a analizar
    PT=PrettyTable()
    PT.title="Numpy array Arithmatic Operations"
    PT.field_names=["Substraction", "Adittion", "Multiplication"]
    PT.add_row([array-arraya, array+arraya, array*arraya]) #Adding row
    print(PT)
```
